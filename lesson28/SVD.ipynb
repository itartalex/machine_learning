```python
import numpy as np
import numpy.linalg as la
import matplotlib.pyplot as plt
```

```python
n=110
a = np.random.random((n,n))
U, s, VT = la.svd(a)
```

```python
Sigma = np.diag(s)
e_r = []
for i in range(2, n):
    r = i
    mat_approx = U[:, :r] @ Sigma[:r, :r] @ VT[:r, :]
    e_r.append(np.linalg.norm(a - mat_approx))
```

```python
print(e_r)
```

```python
plt.plot (e_r, '--');
```
